1. Исследовать на линейную зависимость:$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

т.к. мы можем выразить одну функцию через другую, в данном случае:$$f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)$$следовательно, получаем линейную зависимость

2. Исследовать на линейную зависимость:$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

т.к. мы можем выразить одну функцию через другую, в данном случае:$$f_{4}(x)=f_{3}(x)+2f_{2}(x)+0.5f_{1}(x)$$следовательно, получаем линейную зависимость

3. Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$x = (2, 3, 5) = (2,0,0) + (0,3,0) + (0,0,5) = 1 * (2,0,0) + 3 * (0,1,0) + 0.5 * (0,0,10) = 0.5b_{1} + 1*b_{2} + 3*b_{3} = (0.5, 1, 3)$

4. Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

Для базиса $1$, $x$, $x^{2}$ координаты $(2, -2, 3)$

Для базиса $x^{2}$, $x-1$, $1$ приведем вектор к нужному виду: $3x^{2}-2x+2 = 3x^{2} - 2(x-1)$

тогда координаты для базиса $x^{2}$, $x-1$, $1$ будут $(3, -2, 0)$

5. Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;

б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

а) Возьмем два вектора для примера $a=(0, a2, a3)$ и $b=(b1, 0, b3)$

Сумма данных векторов $a+b=(b1, a2, a3+b3)$.

Ни одна из двух первых координат не равна нулю, следовательно, данная совокупность векторов не является линейным подпространством.

б) Возьмем два вектора для примера $a=2u_{1} + 3u_{2} + 4u_{3}$ и $b=5u_{4} + 6u_{5} + 7u_{6}$

Сумма данных векторов $a+b=(2u_{1}+3u_{2}+4u_{3}+5u_{4}+6u_{5}+7u_{6})$

Умножение на скаляр вектора a: $K=2$, $K*a=4u_{1}+6u_{2}+8u_{3}$

Следовательно, данная совокупность векторов является линейным подпространством.
